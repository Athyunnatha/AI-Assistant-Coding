```python
def count_chars(input_string):
    vowels = 0
    consonants = 0
    digits = 0

    for char in input_string:
        char_lower = char.lower()

        if char_lower.isalpha():
            if char_lower in 'aeiou':
                vowels += 1
            else:
                consonants += 1
        elif char_lower.isdigit():
            digits += 1

    return {"vowels": vowels, "consonants": consonants, "digits": digits}

# Example usage with user input:
user_text = input("Enter a string to analyze: ")
user_counts = count_chars(user_text)
print(f"Original String: '{user_text}'")
print(f"Counts: {user_counts}")
```
